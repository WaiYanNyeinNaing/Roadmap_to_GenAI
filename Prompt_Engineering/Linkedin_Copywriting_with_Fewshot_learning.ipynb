**Load LLM**

```python
from vllm import LLM, SamplingParams
```

```python
llm = LLM(model="mistralai/Mistral-7B-Instruct-v0.1",trust_remote_code=True,dtype="auto")
```

**CopyWriting**

```python
topic = "Prompt Engineering"
target_audience = "busiesss managers, developers, and engineers"
pitch = """
Prompt engineering is the process of structuring text to be interpreted and understood by a generative AI model.
It involves designing inputs, known as prompts, to interact optimally with the AI model, and can include natural language text, images, or other input types.
 Prompt engineering is used to improve the performance of large language models (LLMs) on various tasks such as question answering and arithmetic reasoning.
 It is also important for generating different types of content, including text, images, 3D assets, scripts, and more. Organizations are beginning to hire prompt engineers to optimize the performance of generative AI models, and the practice is expected to become a larger hiring category in the coming years.
"""
```

```python
instruction = f"""
You are a LinkedIn copy writer.
Post length must be less than 3 sentences.

Write a LinkedIn post that simplifies {topic} in your industry, making it easy to understand for your connections.
Target audience — {target_audience}.
Summary of the topic — {pitch}. Topic — {topic}.

"""

print(instruction)
```

```python
sampling_params = SamplingParams(temperature=0.7, top_p=0.95,max_tokens=200)
```

```python
outputs = llm.generate(instruction, sampling_params)

result = outputs[0].outputs[0].text
print(result)
print("")
```
