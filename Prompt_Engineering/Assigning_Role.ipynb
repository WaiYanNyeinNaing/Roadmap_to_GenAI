**Load LLM**

```python
from vllm import LLM, SamplingParams
```

```python
llm = LLM(model="mistralai/Mistral-7B-Instruct-v0.1",trust_remote_code=True,dtype="auto")
```

**Assign Role to AI / Styling Text**

Role prompting is most often used to style text. This involves asking the AI to pretend to be a certain person, or act in a certain way, thus modifying how it writes based on the assigned role. This can be used to change the tone, style, and even the depth of the information presented. Let's delve into this concept with a food review example

```python
topic = "McDonalds"

prompts = f"You are a food critic. Write a review of {topic}. Please write a honest review in conclusion (Good or Bad)."
sampling_params = SamplingParams(temperature=0.8, top_p=0.95,max_tokens=300)
```

```python
outputs = llm.generate(prompts, sampling_params)

result = outputs[0].outputs[0].text
print(result)
```

**Email Writing Example**

Let's consider another example of styling text with role prompting. When using a language model like ChatGPT to draft an email, considering the "role" of the AI is critical to shaping the content it generates. The direction you want your resultant email to take will decide which role you should task the AI with.



```python
topic = "logistical problems"

prompts = f"You are a communications specialist. Draft an email to your client advising them about a delay in the delivery schedule due to {topic}."
sampling_params = SamplingParams(temperature=0.8, top_p=0.95,max_tokens=300)
```

```python
outputs = llm.generate(prompts, sampling_params)

result = outputs[0].outputs[0].text
print(result)
```

**Basic Calculation**

```python
prompts = "What is 100*100/400*56 ?"
sampling_params = SamplingParams(temperature=0.1, top_p=0.95,max_tokens=300)
```

```python
outputs = llm.generate(prompts, sampling_params)

result = outputs[0].outputs[0].text
print(result)
```

**Improving Accuracy**

As mentioned above, the accuracy of the output can be improved with role promptinga. Consider the following example:

```python
prompts = """
You are a brilliant mathematician who can solve any problem in the world.
Attempt to solve the following problem:

What is 100*100/400*56?
"""
sampling_params = SamplingParams(temperature=0.1, top_p=0.95,max_tokens=300)
```

```python
outputs = llm.generate(prompts, sampling_params)

result = outputs[0].outputs[0].text
print(result)
```

**Improving Accuracy with Chain of Thought**

```python
prompts = """
You are a brilliant mathematician who can solve any problem in the world.
Attempt to solve the following problem:

Think step by step:
What is 100*100/400*56?
"""
sampling_params = SamplingParams(temperature=0.1, top_p=0.95,max_tokens=300)
```

```python
outputs = llm.generate(prompts, sampling_params)

result = outputs[0].outputs[0].text
print(result)
```
