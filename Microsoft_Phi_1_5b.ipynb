Build Text Completion Chat with Phi-1.5 Model

````python
import torch
from transformers import AutoModelForCausalLM, AutoTokenizer

torch.set_default_device("cuda")
model = AutoModelForCausalLM.from_pretrained("microsoft/phi-1_5", trust_remote_code=True)
tokenizer = AutoTokenizer.from_pretrained("microsoft/phi-1_5", trust_remote_code=True)
````

**1. Code Completion**

````python
inputs = tokenizer('''```python
def print_prime(n):
   """
   Print all primes between 1 and n
   """, Ans:''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=200)
text = tokenizer.batch_decode(outputs)[0]
print(text)
````

**2. Story Generation**

Generate a story based on a given prompt.

````python
inputs = tokenizer('''Once upon a time in Burma(Myanmar),''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=200)
text = tokenizer.batch_decode(outputs)[0]
print(text)
````

 **3. Dialogue Generation**

Simulate a conversation by providing a line of dialogue.

````python
inputs = tokenizer('''Human: Hello, how are you today?\nAI:,''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=200)
text = tokenizer.batch_decode(outputs)[0]
print(text)
````

**4. Writing Assistance**

Complete an email or a letter.

````python
inputs = tokenizer('''Dear Wai Yan, I am writing to you regarding Machine Learning course,''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=200)
text = tokenizer.batch_decode(outputs)[0]
print(text)
````

**5. Poetry Generation**

Create poetry from a given line or theme

````python
inputs = tokenizer('''Write a poem : The wind whispered through the trees,''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=100)
text = tokenizer.batch_decode(outputs)[0]
print(text)
````

**6. Question Answering**

Answer questions based on context provided.

````python
inputs = tokenizer('''Q: What is the capital of France?\n A:''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=30)
text = tokenizer.batch_decode(outputs)[0]
print(text)
````

**7. Ideas Brainstorming**

Generate ideas on a given topic.

````python
inputs = tokenizer('''Come up with five concepts for mobile apps that help improve mental health and wellbeing: 1''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=200,temperature=1)
text = tokenizer.batch_decode(outputs)[0]
print(text)
````

**8. Philosophical Musings**

Generate philosophical thoughts or musings.

````python
inputs = tokenizer('''Philosophical Musing: What is the meaning of life . .?''', return_tensors="pt", return_attention_mask=False)

outputs = model.generate(**inputs, max_length=100)
text = tokenizer.batch_decode(outputs)[0]
print(text)
````
