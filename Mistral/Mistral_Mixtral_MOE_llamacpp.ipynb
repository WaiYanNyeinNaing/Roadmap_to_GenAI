```python
from llama_cpp import Llama

# Set gpu_layers to the number of layers to offload to GPU. Set to 0 if no GPU acceleration is available on your system.
llm = Llama(
  model_path="./mixtral-8x7b-instruct-v0.1.Q2_K.gguf",  # Download the model file first
  n_ctx=2048,  # The max sequence length to use - note that longer sequence lengths require much more resources
  n_threads=8,            # The number of CPU threads to use, tailor to your system and the resulting performance
  n_gpu_layers=35         # The number of layers to offload to GPU, if you have GPU acceleration available
)

# Simple inference example
output = llm(
  "[INST] {prompt} [/INST]", # Prompt
  max_tokens=512,  # Generate up to 512 tokens
  stop=["</s>"],   # Example stop token - not necessarily correct for this specific model! Please check before using.
  echo=True        # Whether to echo the prompt
)

# Chat Completion API

llm = Llama(model_path="./mixtral-8x7b-instruct-v0.1.Q2_K.gguf", chat_format="llama-2")  # Set chat_format according to the model you are using
llm.create_chat_completion(
    messages = [
        {"role": "system", "content": "You are a story writing assistant."},
        {
            "role": "user",
            "content": "Write a story about llamas."
        }
    ]
)
```

**1. Story Generation**

Generate a story based on a given prompt.

```python
llm.create_chat_completion(
    messages = [
        {"role": "system", "content": "You are a story writing assistant."},
        {
            "role": "user",
            "content": "Write a story about super AI in 2 sentences."
        }
    ]
)
```

 **2. Dialogue Generation**

Simulate a conversation by providing a line of dialogue.

```python
llm.create_chat_completion(
    messages = [
        {"role": "system", "content": "You are helpful AI companion."},
        {
            "role": "user",
            "content": "Human: Hello, how are you doing today?\n AI:"
        }
    ]
)
```

**3. Writing Assistance**

Complete an email or a letter.

```python
llm.create_chat_completion(
    messages = [
        {"role": "system", "content": "You are email writing assistant."},
        {
            "role": "user",
            "content": "Topic: Course Enquiry , Body :Dear Wai Yan, I am writing to you regarding Machine Learning course . ."
        }
    ]
)
```
